```python
# Join all reviews:
review_words = "".join(review for review in df_positive.review)

# Generate a word cloud image based on bag of popcorn
mask = np.array(Image.open("bag.jpg"))

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10, max_font_size = 42,
                max_words=200,
                mask=mask).generate(review_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off") 
plt.tight_layout(pad = 0) 

# store to file
plt.savefig("../Final-Project-IMDB-Reviews/images/positive_wordcloud.png", format="png")

plt.show() 
```
